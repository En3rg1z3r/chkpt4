```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df= pd.read_csv("kc_house_data.csv")
#FIRST LOOK AT THE DATA
print(df.shape)
print(df.dtypes)
df.describe()
df.head()

```

```python
#PLOT
plt.boxplot(df.price)
sns.displot(df.price)
#high positive skewness
df.price = np.log(df.price) 
sns.displot(df.price)
```

```python
#Data-Preprocessing
# days away from 2010/01/01, making the date useless
df["from2013"] = df.date.map(lambda dt: (pd.to_datetime(dt, yearfirst=True)-pd.to_datetime("20130101T000000")).days)
print(df)

#feature selection
#id and zipcode are useless
print(df[df.waterfront==0].shape) #waterfront column is useless
print(df[df.yr_renovated==0].shape) #yr_renovated is useless
print(df[df.view==0].shape) #view is useless too
print(df[df.sqft_basement==0].shape) #this must be kept, 8000~ houses have a basement
df.drop(["date", "zipcode", "waterfront", "yr_renovated","view","id"], axis=1, inplace=True)
df
```

```python
plt.figure(figsize=(16, 6))
hm= sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
hm.set_title('Correlation Heatmap', fontdict={'fontsize':15}, pad=20)
# nearly strong positive corr(and > 0.50) between :
# price & bathrooms
# price & sqft_above
# price & grade
# price & sqft_living
# price & sqft_living15
#these are the main features for the price prediction 

#weak corr : 
# price & bedrooms
# price & lat (probably the nearer to some important places, the higher the price gets )
```

```python
training_data, testing_data = train_test_split(df, test_size=0.2, random_state=25)
```

```python
#linear reg : 
X = training_data[["sqft_living"]]
reg = LinearRegression().fit(X, training_data.price)
print(reg.score(X, training_data.price))

pred= reg.predict(testing_data[["sqft_living"]])
print("MSE : ", metrics.mean_squared_error(testing_data.price, pred))
print("R squared : ", metrics.r2_score(testing_data.price, pred))

Y = reg.coef_* df.sqft_living + reg.intercept_
plt.scatter(df.sqft_living, df.price)
plt.plot(df.sqft_living,Y, 'r')
```

```python
from sklearn.preprocessing import PolynomialFeatures 
x= df[["grade"]]
y= df["price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)  #splitting data
lg=LinearRegression()
poly=PolynomialFeatures(degree=2)

x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)

print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

```python
#multi linear reg
X = training_data[["grade","sqft_living","sqft_above", "bathrooms"]] 
reg = LinearRegression().fit(X, training_data.price)
print(reg.score(X,  training_data.price))

pred= reg.predict(testing_data[["grade","sqft_living", "sqft_above", "bathrooms"]])
print("MSE : ", metrics.mean_squared_error(testing_data.price, pred))
print("R squared : ", metrics.r2_score(testing_data.price, pred))
```
